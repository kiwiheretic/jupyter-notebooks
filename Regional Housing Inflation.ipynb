```python
import pandas as pd
import numpy as np
import string
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
def p2f(x):
    return float(x.strip('%'))

df = pd.read_csv('stats/QV-2017-06-16.csv',converters={'Change value':p2f})
df.head()
```

```python
dfsort = df.sort_values('Change value', axis=0)
dfsort.head()
```

```python
fig, ax = plt.subplots()
plt.rcdefaults()

from pylab import rcParams
rcParams['figure.figsize'] = 10,20

# Example data
x_labels = dfsort['Area']
y_pos = np.arange(len(x_labels))
values = dfsort['Change value']

ax.barh(y_pos, values, align='center',
        color='lightblue', ecolor='black')
ax.set_yticks(y_pos)
ax.set_yticklabels(x_labels)
ax.invert_yaxis()  # labels read top-to-bottom
ax.grid(axis='x', which='major')
ax.set_xlabel('% change')
ax.set_title('NZ Housing Inflation by Region June 2013 - May 2017')

plt.show()
```
