```python
summary1.values
```

```python
#plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
x_labels = summary1.index
y_pos = np.arange(len(x_labels))
values = summary1.values

ax.barh(y_pos, values, align='center',
        color='lightblue', ecolor='black')
ax.set_yticks(y_pos)
ax.set_yticklabels(x_labels)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('# Visas Approved')
ax.set_title('Approved Immigration Work Visas 2015/16')

plt.show()
```
